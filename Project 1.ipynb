```python
# df['A'].replace('A1', 'New_A1', inplace=True) -> example of how to replace values (Found through copilot)
# df.method({col: value}, inplace=True) or df[col].methid(value) -> suggestions given by python/vscode


# cl["CostOfLivingIndexViaNumbeo_2025"].replace('NaN', '38.75', inplace=True) -> didn't work - syntax of inplace was removed in python 3
# df.method({col: value}, inplace=True) or df[col].methid(value) -> suggestions given by python/vscode

# cl["CostOfLivingIndexViaNumbeo_2025"].replace("NaN", 38.75) -> didn't work maybe did something wrong

# cl.replace({"CostOfLivingIndexViaNumbeo_2025": 38.75}, inplace=True)

# I found a working code! this is from geeksforgeeks
cl["CostOfLivingIndexViaNumbeo_2025"] = cl["CostOfLivingIndexViaNumbeo_2025"].fillna(cl["CostOfLivingIndexViaNumbeo_2025"].median())
```
